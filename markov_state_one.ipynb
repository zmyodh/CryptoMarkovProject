```python
import numpy as np
import pandas as pd
df = pd.read_csv('solana_data_mod.csv', sep='\t', index_col=0)
```

```python
df['State'] = np.where(df['Price'].diff() > 0, 'up', 'down')
```

```python
# Total Number of Previous States that were up or down, which is just the count of all Up states or Down states
total_up = (df["State"] == "up").sum()
total_down = (df["State"] == "down").sum()
if (len(df["State"]) == total_up + total_down):
    print("Addition matches")
else:
    print(f"length of states: {len(df['State'])} <--> Added total_up + total_down: {total_up + total_down}")
# Number of times state was up and previous state was up
up_up_days = ((df["State"] == "up") & (df["State"].shift(1) == "up")).sum()

# Number of times state was up and previous state was down
up_down_days = ((df["State"] == "up") & (df["State"].shift(1) == "down")).sum()

# Number of times state was down and previous state was up
down_up_days = ((df["State"] == "down") & (df["State"].shift(1) == "up")).sum()

# Number of times state was down and previous state was down
down_down_days = ((df["State"] == "down") & (df["State"].shift(1) == "down")).sum()


print(f"Number of consecutive up-up days: {up_up_days}")
print(f"Number of consecutive up-down days: {up_down_days}")
print(f"Number of consecutive down-up days: {down_up_days}")
print(f"Number of consecutive down-down days: {down_down_days}")
print(f"Number of up states: {total_up}")
print(f"Number of up states: {total_down}")
```

```python
# say that we want to know do the same for different states 
# lets say you want to know when the equity was up -> up -> ?
'''
P(up-up-up), P(up-up-down)
P(up-down-up), P(up-down-down)
P(down-up-up), P(down-up-down)
P(down-down-up), P(down-down-down)

Remember that we have a larger matrix now because we increased n by 1 so 2^n where n=3 ==> 8

and keep in mind we need total amount we find up-up-up / total amount of times we count up-up for instance P(up-up-up)
'''
```

```python
# create transition matrix
up_up_ratio = round(up_up_days/total_up, 2)
up_down_ratio = round(up_down_days/total_up, 2)
down_up_ratio = round(down_up_days/total_down, 2)
down_down_ratio = round(down_down_days/total_down, 2)

print(f"up-up ratio: {up_up_ratio}")
print(f"up-down ratio: {up_down_ratio}")
print(f"down-up ratio: {down_up_ratio}")
print(f"down-down ratio: {down_down_ratio}")
```

```python
transition_matrix = pd.DataFrame(
    {"up": [up_up_ratio, down_up_ratio] 
     ,"down": [up_down_ratio, down_down_ratio]}, index = ["up", "down"])
# index are the rows... which can sometimes be confusing because it looks like it's in row format when you code it,
# but it is actually in column format.  think of it was "up" : [x, y] as a column vector
```

```python
print(transition_matrix)
```

```python
# Lets do it the silly way first 

P = transition_matrix.to_numpy()
print("Transition Matrix P = \n",P)

one = np.linalg.matrix_power(P, 1)
two = np.linalg.matrix_power(P, 2)
five = np.linalg.matrix_power(P, 5)
twenty = np.linalg.matrix_power(P, 20)
one_hundo = np.linalg.matrix_power(P, 100)

pi = np.array([0,1]) # this means it was a down day in previous state

print(f"power of 1:\n  {one}")
print(f"power of 2:\n  {two}")
print(f"power of 5:\n  {five}")
print(f"power of 20:\n  {twenty}")
print(f"power of 100:\n  {one_hundo}")
```

```python
import scipy.linalg
values, left_vectors = scipy.linalg.eig(P, right=False, left=True)

print("EigenValues: ", values)
print("EigneVectors:\n ", left_vectors)
```

```python
pi = left_vectors[:,0]
pi_normalized = [(x/np.sum(pi)).real for x in pi]
pi_normalized
```

So now you might be asking... well what do these numbers mean.... well thats a great question

It means that in any random walk, wherever we start, we end with these two probabilities.

So if we start at "UP" [1,0] we take a random walk, our equilibrium lambda = 1, will give you those probabilies. from any initial state if up next day we have ~52% chance it will be up and ~48% itll be down
