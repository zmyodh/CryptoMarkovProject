```python
import json
import numpy as np
import pandas as pd
with open("sol_data.json", "r") as read_file:
    sol_data = json.load(read_file)
```

```python
prices_df = pd.DataFrame(sol_data["prices"], columns=["Timestamp", "Price"])
marketcap_df = pd.DataFrame(sol_data["market_caps"], columns=["Timestamp", "MarketCap"])
total_volumes_df = pd.DataFrame(sol_data["total_volumes"], columns=["Timestamp", "TotalVolumes"])

df = pd.merge(prices_df, marketcap_df, on="Timestamp")
df = pd.merge(df, total_volumes_df, on="Timestamp")
```

```python
import math

def shift_time(number, power, base=10):
    return number * base**power

power = -3
TimestampMod = []
for time in df['Timestamp']:
    #print(time, round(shift_time(time, power), 3)) # new type is a float
    time_mod = round(shift_time(time, power), 3)
    TimestampMod.append(round(shift_time(time, power), 3))
```

```python
TimestampMod
```

```python
df['TimestampMod'] = TimestampMod
```

```python
# convert unix times to readable dates
df['date'] = pd.to_datetime(df['TimestampMod'], unit='s')

# extract year, month, day, hour, minute, and second
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df['second'] = df['date'].dt.second
```

np.where is a NumPy function that allows you to create a new array or Series by choosing elements from x or y based on a specified condition.
It works as follows:
    If the condition is True, it selects the corresponding element from x.

```python
# Total Number of Previous States that were up or down, which is just the count of all Up states or Down states
total_up = (df["State"] == "up").sum()
total_down = (df["State"] == "down").sum()
if (len(df["State"]) == total_up + total_down):
    print("Addition matches")
else:
    print(f"length of states: {len(df['State'])} <--> Added total_up + total_down: {total_up + total_down}")
# Number of times state was up and previous state was up
up_up_days = ((df["State"] == "up") & (df["State"].shift(1) == "up")).sum()

# Number of times state was up and previous state was down
up_down_days = ((df["State"] == "up") & (df["State"].shift(1) == "down")).sum()

# Number of times state was down and previous state was up
down_up_days = ((df["State"] == "down") & (df["State"].shift(1) == "up")).sum()

# Number of times state was down and previous state was down
down_down_days = ((df["State"] == "down") & (df["State"].shift(1) == "down")).sum()


print(f"Number of consecutive up-up days: {up_up_days}")
print(f"Number of consecutive up-down days: {up_down_days}")
print(f"Number of consecutive down-up days: {down_up_days}")
print(f"Number of consecutive down-down days: {down_down_days}")
print(f"Number of up states: {total_up}")
print(f"Number of up states: {total_down}")

```

```python
# create transition matrix
up_up_ratio = up_up_days/total_up
up_down_ratio = up_down_days/total_up
down_up_ratio = down_up_days/total_down
down_down_ratio = down_down_days/total_down

print(f"up-up ratio: {up_up_ratio}")
print(f"up-down ratio: {up_down_ratio}")
print(f"down-up ratio: {down_up_ratio}")
print(f"down-down ratio: {down_down_ratio}")
```

```python
transition_matrix = pd.DataFrame(
    {"up": [up_up_ratio, down_up_ratio] 
     ,"down": [up_down_ratio, down_down_ratio]}, index = ["up", "down"])
# index are the rows... which can sometimes be confusing because it looks like it's in row format when you code it,
# but it is actually in column format.  think of it was "up" : [x, y] as a column vector
```
