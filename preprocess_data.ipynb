```python
import json
with open("solana_data.json", "r") as read_file:
    readmebitch = json.load(read_file)
```

```python
import pandas as pd
prices_df = pd.DataFrame(readmebitch["prices"], columns=["Timestamp", "Price"])
marketcap_df = pd.DataFrame(readmebitch["market_caps"], columns=["Timestamp", "MarketCap"])
total_volumes_df = pd.DataFrame(readmebitch["total_volumes"], columns=["Timestamp", "TotalVolumes"])

df = pd.merge(prices_df, marketcap_df, on="Timestamp")
df = pd.merge(df, total_volumes_df, on="Timestamp")
```

```python
df.head()
```

```python
import math

def shift_time(number, power, base=10):
    return number * base**power

power = -3
TimestampMod = []
for time in df['Timestamp']:
    #print(time, round(shift_time(time, power), 3)) # new type is a float
    time_mod = round(shift_time(time, power), 3)
    TimestampMod.append(round(shift_time(time, power), 3))
```

```python
df['TimestampMod'] = TimestampMod
```

```python
# convert unix times to readable dates
df['date'] = pd.to_datetime(df['TimestampMod'], unit='s')

# extract year, month, day, hour, minute, and second
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df['second'] = df['date'].dt.second
```

```python
df.to_csv('solana_data_mod.csv', sep='\t', encoding= 'utf-8')
```
